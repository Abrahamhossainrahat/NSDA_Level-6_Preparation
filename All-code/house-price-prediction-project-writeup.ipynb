# 🏡 House Price Prediction — Project Writeup

## 📌 Overview  
This project focuses on predicting house prices using various property features such as area, number of bedrooms, bathrooms, stories, parking availability, and main road access. The goal is to build a reliable regression model that can estimate property prices based on these features.

## 📊 Dataset Description  
The dataset contains essential information about residential properties, including:  
- **Area (sq ft)**  
- **Number of bedrooms**  
- **Number of bathrooms**  
- **Stories**  
- **Parking availability**  
- **Main road access (Yes/No)**  
- **Target variable: Price**

This dataset helps identify how different features influence the final market price.

## 🔍 Objective  
The main objective of this project is to:  
- Understand the relationship between property features and price  
- Build a predictive model using Linear Regression  
- Evaluate the performance of the model using appropriate metrics  
- Gain insights for both buyers and sellers in the housing market

## 🛠 Approach / Methodology  
### 1. Exploratory Data Analysis (EDA)  
- Checked data distribution  
- Looked for correlations  
- Identified important features  
- Converted categorical features using One-Hot Encoding

### 2. Data Preprocessing  
- Removed missing values  
- Converted categorical columns to numeric  
- Normalized/Scaled the data if necessary  

### 3. Model Building  
- Applied Linear Regression as the baseline model  
- Split the data into Train and Test sets  
- Trained the model on the training set  

### 4. Model Evaluation  
Evaluation metrics used:  
- R² Score  
- MAE  
- MSE  
- RMSE  

## 📈 Results  
The Linear Regression model performed well and revealed:  
- Larger area increases price  
- More bedrooms and bathrooms lead to higher value  
- Main road access increases price  
- Parking availability adds value  

This model provides a strong baseline for further improvement using more advanced algorithms.

## ✅ Conclusion  
This project demonstrates how property features influence house prices and how a simple Linear Regression model can be used to predict housing costs. With advanced models and improved feature engineering, prediction accuracy can be increased further.
